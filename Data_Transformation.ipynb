```python
import pandas as pd

df = pd.read_csv("/content/employee_productivity_dataset.csv")
df.head()
```

Now, let's handle the missing values in the `df` DataFrame according to your specifications:
- **Age**: Fill with Median
- **Salary**: Fill with Mean
- **Hours_Worked_Per_Week**: Fill with Median
- **Performance_Score**: Fill with Mean

```python
# Calculate median for 'Age' and 'Hours_Worked_Per_Week'
median_age = df['Age'].median()
median_hours_worked = df['Hours_Worked_Per_Week'].median()

# Calculate mean for 'Salary' and 'Performance_Score'
mean_salary = df['Salary'].mean()
mean_performance_score = df['Performance_Score'].mean()

# Fill missing values
df['Age'].fillna(median_age, inplace=True)
df['Salary'].fillna(mean_salary, inplace=True)
df['Hours_Worked_Per_Week'].fillna(median_hours_worked, inplace=True)
df['Performance_Score'].fillna(mean_performance_score, inplace=True)

print("DataFrame after handling missing values:")
display(df.head())
```

Q4. Normalization (Min-Max Scaling)
Normalize the following columns:
Salary
Hours_Worked_Per_Week
Use MinMaxScaler and display results.

Min-Max Scaling (Normalization) transforms features by scaling each feature to a given range, typically [0, 1]. This ensures that all features contribute equally to the distance calculations, which is important for many machine learning algorithms. The formula is:

$X_{norm} = (X - X_{min}) / (X_{max} - X_{min})$

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max Scaling to 'Salary' and 'Hours_Worked_Per_Week'
# We will create new columns for the normalized values
df_encoded['Salary_Normalized'] = scaler.fit_transform(df_encoded[['Salary']])
df_encoded['Hours_Worked_Per_Week_Normalized'] = scaler.fit_transform(df_encoded[['Hours_Worked_Per_Week']])

print("DataFrame with normalized 'Salary' and 'Hours_Worked_Per_Week' columns:")
display(df_encoded[['Salary', 'Salary_Normalized', 'Hours_Worked_Per_Week', 'Hours_Worked_Per_Week_Normalized']].head())
```

Q5. Standardization (Scaling)
Apply StandardScaler on:
Age Projects_Completed
Display the transformed values.

Standardization (often called Z-score normalization) transforms data such that it has a mean of 0 and a standard deviation of 1. This scaling technique is particularly useful for algorithms that assume a Gaussian distribution of the features, like linear regression, logistic regression, and linear discriminant analysis.

$X_{scaled} = (X - \mu) / \sigma$

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to 'Age' and 'Projects_Completed'
# We will create new columns for the standardized values
df_encoded['Age_Scaled'] = scaler.fit_transform(df_encoded[['Age']])
df_encoded['Projects_Completed_Scaled'] = scaler.fit_transform(df_encoded[['Projects_Completed']])

print("DataFrame with standardized 'Age' and 'Projects_Completed' columns:")
display(df_encoded[['Age', 'Age_Scaled', 'Projects_Completed', 'Projects_Completed_Scaled']].head())
```

Q7. Build Preprocessing Pipeline
Create a pipeline that:
Applies encoding to categorical columns
Applies scaling to numerical columns
Use:
ColumnTransformer
Pipeline

A preprocessing pipeline combines multiple data transformations into a single object, ensuring that the same steps are applied consistently to your data. `ColumnTransformer` allows us to apply different transformations to different subsets of columns, while `Pipeline` chains these transformations together, often culminating in an estimator.

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
import pandas as pd

# Define categorical and numerical features based on previous steps
categorical_features = ['Gender', 'Department', 'Work_Mode', 'Location']
numerical_features_minmax = ['Salary', 'Hours_Worked_Per_Week']
numerical_features_standard = ['Age', 'Projects_Completed', 'Performance_Score']

# Create preprocessing steps using Pipelines for each type of transformation
# Categorical transformer: OneHotEncoder
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Numerical transformer for Min-Max Scaling
numerical_transformer_minmax = Pipeline(steps=[
    ('minmax', MinMaxScaler())
])

# Numerical transformer for Standard Scaling
numerical_transformer_standard = Pipeline(steps=[
    ('standard', StandardScaler())
])

# Create a ColumnTransformer to apply different transformations to different columns
# 'remainder="drop"' will drop columns not specified (like Employee_ID, Name, Join_Date)
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features),
        ('num_minmax', numerical_transformer_minmax, numerical_features_minmax),
        ('num_standard', numerical_transformer_standard, numerical_features_standard)
    ],
    remainder='drop'
)

# Create the full preprocessing pipeline
# The final step 'preprocessor' is the ColumnTransformer itself, as we only need preprocessing here.
preprocessing_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Fit and transform the data using the pipeline
# We use the original `df` which has missing values imputed but no encoding/scaling applied to original columns yet
df_processed_array = preprocessing_pipeline.fit_transform(df)

# Get the names of the transformed features for creating a DataFrame
transformed_feature_names = []

# Get names for one-hot encoded features
cat_feature_names = preprocessing_pipeline.named_steps['preprocessor'].named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_features)
transformed_feature_names.extend(cat_feature_names)

# Get names for Min-Max scaled numerical features
transformed_feature_names.extend(numerical_features_minmax)

# Get names for Standard scaled numerical features
transformed_feature_names.extend(numerical_features_standard)

# Convert the processed array back to a DataFrame for readability
df_transformed = pd.DataFrame(df_processed_array, columns=transformed_feature_names)

print("Shape of the preprocessed data:", df_transformed.shape)
print("\nFirst 5 rows of the transformed DataFrame:")
display(df_transformed.head())
```

Q9. Conceptual Question
Why is scaling important in Python?

Scaling data is crucial in many machine learning tasks in Python for several reasons:

Equal Contribution: Many machine learning algorithms (like Support Vector Machines, K-Nearest Neighbors, and neural networks) calculate distances between data points. If features have vastly different scales, the feature with the larger scale will dominate the distance calculation, making the smaller-scaled features almost irrelevant. Scaling ensures all features contribute equally.

Algorithm Performance: Some algorithms converge faster or perform better when features are on a similar scale. For instance, gradient descent-based algorithms (like those used in linear regression or neural networks) benefit from scaling as it helps avoid oscillations and speeds up convergence.

Preventing Domination by Outliers: Features with larger ranges are more susceptible to the influence of outliers. Scaling can help to mitigate this by bringing all features into a comparable range.

Interpretability: While not directly related to model performance, scaling can sometimes make coefficients or feature importances more interpretable, especially when comparing the impact of different features.

Regularization: Techniques like L1 or L2 regularization (Lasso and Ridge regression) penalize large coefficients. If features are not scaled, features with larger magnitudes might be unfairly penalized more heavily.

Q10. Conceptual Question
Why do we convert categorical data into numerical form?

Converting categorical data into numerical form is essential in machine learning for several key reasons:

Algorithm Compatibility: Most machine learning algorithms are designed to work with numerical input. They rely on mathematical operations (like calculating distances, gradients, or weights) that are not directly applicable to text-based or categorical labels. By converting categories to numbers, we make the data compatible with these algorithms.

Order and Magnitude: While some categorical features might have an inherent order (e.g., 'low', 'medium', 'high'), many do not (e.g., 'red', 'green', 'blue'). Numerical representations allow algorithms to understand and utilize this order (in the case of ordinal encoding) or treat each category distinctly without imposing an artificial order (in the case of one-hot encoding).

Feature Engineering: Numerical representations open up possibilities for further feature engineering. For example, with one-hot encoded features, you can create interaction terms.

Avoiding Misinterpretation: If categorical data were used directly as strings, an algorithm might treat them as arbitrary text rather than distinct categories, leading to errors or poor model performance. Even if they are mapped to arbitrary numbers without proper encoding, an algorithm might infer an unintended ordinal relationship (e.g., thinking 'category 2' is 'greater' than 'category 1').
